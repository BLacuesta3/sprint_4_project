Code Meaning:  Use the isna() and sum() functions in order to count the number of missing values in the dataframe.

Code Result:  There seems to be a significant amount of missing values in the dataframe.  The values should
be renamed instead of deleted in order not to lose any significant information.

```python
df.isna().sum()
```

```python
df.info()
```

```python
df['cylinders'] = df[['cylinders', 'type']].groupby('type').transform(lambda x:x.fillna(x.median()))

df['cylinders'].isna().sum()
```

```python
df['odometer'] = df.groupby('type')['odometer'].transform(lambda x: x.fillna(x.mean()))

df['odometer'].isna().sum()
```

```python
df['model_year'] = df.groupby('model')['model_year'].transform(lambda x: x.fillna(x.mode().iloc[0]))

df['model_year'].isna().sum()
```

```python
#set = unique values
paint_color_list = df[~df['paint_color'].isna()]['paint_color'].to_list()

paint_color_list
```

```python
df['paint_color'] = df['paint_color'].fillna(random.choice(paint_color_list))

df['paint_color'].isna().sum()
```

```python
is_4wd_list = df[~df['is_4wd'].isna()]['is_4wd'].to_list

is_4wd_list
```

```python
df['is_4wd'] = df['is_4wd'].fillna(random.choice(is_4wd_list))

df['is_4wd'].isna().sum()
```

Code Meaning: Create a scatter plot using px.scatter() in order to show how car price and car quality affect the number of days the car is listed on the lot. The days that the car was listed indicates how long it took to sell the car upon arrival.  
The x axis indicates the  price of each car. The y axis indicates the number of days each car is listed. The conditions of each car are color coded.

Code Result:  It appears that cars priced 50k and below sell faster than more expenisve cars.

```python
fig_scatter_1 = px.scatter(df, x= 'price', y='days_listed',color='condition', title="Relationship Between Car Price/ Car Quality and Days Listed", labels={'price': 'Car Price', 'days_listed': 'Number of Days Listed'})

fig_scatter_1.show()
```
